```python
weather = pd.read_csv('./weather.csv')
```

```python
weather #whenever records are duplicated, it is must likely variables are stored in a row
```

```python
#you can use pd.melt or weather.melt
weather_melt = pd.melt(
    weather,
    id_vars = ['id','year','month', 'element'],
    var_name = 'day',
    value_name = 'temp'
)
```

```python
weather_melt.head() 

#we need to separate the tmax and tmin- if we want to fit a model based of year, month and day also temperaturize your response.
```

```python
#using a PIVOT TABLE
```

```python
weather_melt
```

```python
weather_tidy = weather_melt.pivot_table(
    index=['id','year','month','day'],
    columns= 'element',
    values= 'temp'
)   
#we specify the column to pivot and the values to populate the columns. The Null values are removed in PIVOT TABLE
```

```python
weather_tidy
```

```python
weather_tidy.reset_index() #to work with a flat df use the pd.reset_index()
```
